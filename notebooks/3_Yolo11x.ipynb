```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
import os
import zipfile

# Load model
model = YOLO('/content/best.pt')

# Open video
video_path = '/content/vid.mp4'
cap = cv2.VideoCapture(video_path)

# Settings
frame_interval = 10  # Process every 10th frame
max_frames = 6       # Only 6 frames
processed = 0
frame_count = 0

# Output directory for frames
output_dir = "/content/annotated_frames"
os.makedirs(output_dir, exist_ok=True)

while cap.isOpened() and processed < max_frames:
    success, frame = cap.read()
    if not success:
        break

    if frame_count % frame_interval == 0:
        results = model(frame)
        annotated_frame = results[0].plot()

        # Save annotated frame
        output_path = f"{output_dir}/frame_{processed+1}.jpg"
        cv2.imwrite(output_path, annotated_frame)

        # Show (optional)
        cv2_imshow(annotated_frame)

        processed += 1

    frame_count += 1

cap.release()

# Zip the saved frames
zip_path = "/content/annotated_frames.zip"
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for filename in os.listdir(output_dir):
        zipf.write(os.path.join(output_dir, filename),
                   arcname=filename)

# Download zip file
from google.colab import files
files.download(zip_path)
```

```python
from ultralytics import YOLO
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Load the model
model = YOLO('/content/best.pt')

# Run detection
results = model('/content/lonecheetah.jpg')[0]

# Draw bounding boxes manually in dark green with larger white font
img = results.orig_img.copy()

for box in results.boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    conf = float(box.conf[0])
    cls = int(box.cls[0])
    label = f"{model.names[cls]} {conf:.2f}"

    # Draw dark green box
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 100, 0), 3)

    # Draw label in white with bigger font
    font_scale = 1.2
    thickness = 2
    (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, font_scale, thickness)
    cv2.rectangle(img, (x1, y1 - h - 10), (x1 + w, y1), (0, 100, 0), -1)  # Label background
    cv2.putText(img, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), thickness)

# Show the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
from ultralytics import YOLO
import cv2
import time
import numpy as np

# Load model
model = YOLO('/content/best.pt')  # Use your trained model

# Open 30-second video
video_path = '/content/30.mp4'
cap = cv2.VideoCapture(video_path)

# Get video FPS and total frame count
fps = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
duration = total_frames / fps

print(f"Video Duration: {duration:.2f} seconds | FPS: {fps:.2f} | Total Frames: {total_frames}")

# Read frames and time inference
detection_times = []
frame_count = 0

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Stop if frame count crosses 30s worth of frames
    if frame_count >= int(30 * fps):
        break

    start = time.time()
    _ = model(frame)
    end = time.time()

    detection_times.append((end - start) * 1000)  # milliseconds
    frame_count += 1

cap.release()

# Mean detection time
mean_dt = np.mean(detection_times)
print(f"Processed {frame_count} frames")
print(f"Mean Detection Time for 30s video: {mean_dt:.2f} ms per frame")
```

```python
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
import cv2
from google.colab.patches import cv2_imshow

# Load your custom YOLO model
model = YOLO("yolo11x.pt")  # Replace with your actual path

# Initialize DeepSORT tracker
tracker = DeepSort(max_age=30, min_hits=3, iou_threshold=0.3)

# Load video
video_path = "/content/3.mp4"  # Replace with actual video file
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Output video writer (as MP4)
output_path = "output.mp4"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run detection using YOLO
    results = model(frame)[0]

    detections = []
    for r in results.boxes:
        x1, y1, x2, y2 = map(int, r.xyxy[0])
        conf = float(r.conf)  # Confidence score from YOLO
        cls = int(r.cls)
        # Add bounding box, confidence score, and class to detections
        detections.append(([x1, y1, x2 - x1, y2 - y1], conf, cls))

    # Update tracker with new detections
    tracks = tracker.update_tracks(detections, frame=frame)

    # Process each tracked object
    for track in tracks:
        if not track.is_confirmed():
            continue

        track_id = track.track_id
        ltrb = track.to_ltrb()
        x1, y1, x2, y2 = map(int, ltrb)

        # Draw bounding box around the tracked object
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Add text with track ID and confidence score from YOLO
        label = f"Cheetah {track_id} | {r.conf*100:.2f}%"
        cv2.putText(frame, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Write frame to output video
    out.write(frame)

    # Optional: Display the frame (for testing in Colab)
    if frame_count < 3:
        cv2_imshow(frame)

    frame_count += 1
    if frame_count == 100:  # Optional limit for demo purposes
        break

cap.release()
out.release()
```
